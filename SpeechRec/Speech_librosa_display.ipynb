```python
# pip install librosa
import librosa

# 读取音频
y, sr = librosa.load('music1.wav')
print(y)
print(sr)
```

```python
# 修改采样率
y, sr = librosa.load('music1.wav', sr=16000)
print("sr=16000: ", sr)
y, sr = librosa.load('music1.wav', sr=None)
print("sr=None: ", sr)
```

```python
# 提取特征

# Log-Mel Spectrogram 特征
melspec = librosa.feature.melspectrogram(y, sr, n_fft=1024, hop_length=512, n_mels=128)
logmelspec = librosa.power_to_db(melspec)
print(logmelspec.shape)
print(logmelspec)
```

```python
# MFCC特征
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40)
print(mfccs.shape)
print(mfccs)
```

```python
# 绘图显示
import librosa.display
import matplotlib.pyplot as plt

# 绘制声音波形
plt.figure()
librosa.display.waveplot(y, sr)
plt.title('Music waveform')
plt.show()
```

```python
# 绘制频谱图
melspec = librosa.feature.melspectrogram(y, sr, n_fft=1024, hop_length=512, n_mels=128)
logmelspec = librosa.power_to_db(melspec)
plt.figure()
librosa.display.specshow(logmelspec, sr=sr, x_axis='time', y_axis='mel')
plt.title('Music waveform')
plt.show()
```
